# Implement GreedySorting to Sort a Permutation by Reversals

```python
import numpy as np
```

```python
task_file = open('rosalind_ba6a.txt','r')
task = ''.join(task_file).replace('(', '').replace(')', '').split()

P = [int(i) for i in task]
```

```python
def greedySort(P):
    apDist = []
    for k in range(len(P)):
        while P[k] != k + 1:
            kSort(P, k)
            apDist.append(list(P))
    output = ''

    for i in apDist:
        output += '(' + ' '.join(['+' + str(x) if x > 0 else str(x) for x in i]) + ')\n'
    return ''.join(output)


def kSort(P, k):
    j = k
    while P[j] != k + 1 and P[j] != -(k + 1):
        j += 1
    P[k:j+1] = [-x for x in P[k:j+1][::-1]]
```

```python
print(greedySort(P))
```
